```python
# Parameters
series_id = "2456"
```

```python
arena_time_df = pd.read_csv('arena_time_df_'+series_id+'.csv')
```

```python
avg_arena_times = arena_time_df.groupby('Arena').mean().reset_index()
```

```python
overall_avg_time = avg_arena_times['adj_time'].mean()
avg_arena_times = avg_arena_times.append({'Arena' : 'AVERAGE', 'adj_time' : overall_avg_time},ignore_index=True,)
avg_arena_times.sort_values('adj_time', ascending = False,inplace=True)
```

```python
#Sorting by median time
arena_time_df = arena_time_df.join(arena_time_df.groupby('Arena').median(), on='Arena', rsuffix='_median')
arena_time_df.sort_values('adj_time_median', inplace=True)
```

```python
# probably filtering outliers
## very short games occassionally show up
arena_time_df = arena_time_df[arena_time_df.adj_time > 120]
```

```python
import plotly.graph_objs as go
```

```python
arena_time_df.adj_time = pd.to_datetime(arena_time_df.adj_time,unit='s')
```

```python
data = [
    go.Box(
        x = arena_time_df['adj_time'],
        y = arena_time_df['Arena'],
        boxpoints='suspectedoutliers',
        orientation = 'h'
    )
]
```

```python
layout = go.Layout(
    title='Boxplot of Arena Times - MFP Series_id'+series_id,
    xaxis=dict(
        title='Arena Time (min:sec)',
        titlefont=dict(
            family='Courier New, monospace',
            size=18,
            color='#7f7f7f'
        ),
        dtick=86400000.0*(2.5/1440),
        tick0="1/1/1970",
        tickformat="%M:%S"
    ),
    yaxis=dict(
        title='Arena',
        titlefont=dict(
            family='Courier New, monospace',
            size=12,
            color='#7f7f7f'
        )
    ),
    template = 'seaborn'
)
```

```python
fig = go.Figure(data=data, layout=layout)
```
