```python
group_round_times = df.groupby(['round_number','group_number']).mean().reset_index()
data = [
    go.Heatmap(
        x=group_round_times['round_number'],
        y=group_round_times['group_number'],
        z=group_round_times['adj_time']
    )
]
layout = go.Layout(
    yaxis=dict(
        autorange='reversed'
    )
)
fig = go.Figure(data=data, layout=layout)
py.iplot(fig, filename='round-group-time-heatmap')
```
