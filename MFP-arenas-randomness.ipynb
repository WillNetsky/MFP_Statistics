```python
import pandas as pd
import numpy as np
import io
import requests
import json
```

```python
# Matchplay API URL
series_id = '748'
series_url = 'https://matchplay.events/data/series/'+series_id+'/standings'
```

```python
# Call the API
r = requests.get(series_url)
data = r.json()
tournaments = data['tournament_points']
tournament_ids = []
for week in tournaments:
    if week['status'] != 'planned':
        tournament_ids.append(week['tournament_id'])
```

```python
tournaments = []
for tournament in tournament_ids:
    results_url = 'https://matchplay.events/data/tournaments/' + str(tournament) + '/results/csv'
    results = pd.read_csv(results_url)
    tournaments.append(results)
results = pd.concat(tournaments)
```

```python
arenas = results[['Round','Set','Game 1']]
```

```python
arenas.drop_duplicates(inplace = True)
```

```python
arena_counts = arenas['Game 1'].value_counts()
```

```python
arena_counts/sum(arena_counts)*100
```
