```python
def multiple_season_pull_apply(row):
    params = dict(series_id = str(row.series_id))
    print("Running series_id: "+str(row.series_id))
    pm.execute_notebook(
       'MFP-arenas-times.ipynb',
       'output/MFP-arenas-times'+today.strftime('%m-%m-%y-%H:%M')+'.ipynb',
       parameters=params
    )
    pm.execute_notebook(
       'MFP-arenas-plot.ipynb',
       'output/MFP-arenas-plot'+today.strftime('%m-%m-%y-%H:%M')+'.ipynb',
       parameters=params
    )
    pm.execute_notebook(
       'mfp_standings_projections.ipynb',
       'output/mfp_standings_projections'+today.strftime('%m-%m-%y-%H:%M')+'.ipynb',
       parameters=params
    )
    pm.execute_notebook(
       'mfp_standings_to_html.ipynb',
       'output/mfp_standings_to_html'+today.strftime('%m-%m-%y-%H:%M')+'.ipynb',
       parameters=params
    )
    
    clear_output(wait=True)
    clear_output(wait=True)
    return row
```
