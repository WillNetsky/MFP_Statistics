# MFP All Seasons
The goal of this notebook is to gather all seasons of MFP into a single dataframe

```python
import pandas as pd
import os
```

```python
import pickle

season_id_dict = {
    '209' : 'Winter 17',
    '285' : 'Spring 17',
    '350' : 'Summer 17',
    '409' : 'Fall 17',
    '475' : 'Winter 18',
    '582' : 'Spring 18',
    '672' : 'Summer 18',
    '748' : 'Fall 18',
    '861' : 'Winter 19',
    '987' : 'Spring 19',
    '1080': 'Summer 19',
    '1194': 'Fall 19',
    '1357': 'Winter 20',
    '1755': 'Fall 21',
    '1875': 'Winter 22',
    '2049': 'Spring 22',
    '2149': 'Summer 22',
    '2250': 'Fall 22',
    '2456': 'Winter 23',
    '2607': 'Spring 23'
}

with open('season_id_dict.pickle', 'wb') as f:
    pickle.dump(season_id_dict, f)
```

```python
season_id_dict.keys()
```

```python
season_dfs = []
for season in season_id_dict.keys():
    temp = pd.read_csv('seasons/mfp_league_'+season+'_standings.csv')
    new_names = temp.columns.values
    new_names[0] = 'name'
    for i in range(1,temp.filter(regex='^\d+$').shape[1]+1):
        new_names[i] = 'week'+str(i)
    temp.columns = new_names
    temp['season_id'] = season
    season_dfs.append(temp)
```

```python
mfp_all_seasons = pd.concat(season_dfs, sort=True)
```

```python
mfp_all_seasons.columns
```

```python
mfp_all_seasons.to_csv('mfp_all_seasons.csv',index=False)
```
