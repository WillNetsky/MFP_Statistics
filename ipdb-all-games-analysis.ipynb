```python
from bs4 import BeautifulSoup
import pandas as pd
```

```python
f = open('ipdb-alpha-game-list.html', 'r')
page = f.read()
f.close()
```

```python
soup = BeautifulSoup(page)
```

```python
table = soup.find_all('table')[0]
```

```python
df = pd.read_html(str(table),header=0)[0]
```

```python
df.dropna(inplace=True)
```

```python
df['Date Manufactured'] = pd.to_datetime(df['Date Manufactured'])
```

```python
df = df[df['Date Manufactured'] >= '1960-01-01']
```

```python
df.sort_values('Date Manufactured')
```

```python
df = df.groupby('Theme').filter(lambda x: len(x) < 10 and len(x) > 2)
```

```python
df.Theme.value_counts()
```
