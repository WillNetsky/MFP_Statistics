```python
import pandas as pd
import os
```

```python
season_files = os.listdir('/Users/willn/Dropbox/Code/MFP_Statistics/seasons_full')
```

```python
season_ids = []
for season in season_files:
    season_ids.append(season.split('_')[2])
season_ids
```

```python
season_dfs = []
for season in season_ids:
    temp = pd.read_csv('seasons_full/mfp_league_'+season+'_full.csv')
    temp['season_id'] = season
    season_dfs.append(temp)
```

```python
mfp_all_seasons_full = pd.concat(season_dfs, sort=True)
```

```python
mfp_all_seasons_full.columns
```

```python
mfp_all_seasons_full.to_csv('mfp_all_seasons_full.csv',index=False)
```
